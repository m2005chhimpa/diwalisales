```python
##########################################
ax = sns.countplot(x='Gender' , data =df)
plt.title('Total Male and Female')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
```

# most of the buyer are female and even purchasing of female > male

```python
ax = sns.countplot(x='Age Group', data=df, hue='Gender')

# Add labels to each bar
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Count of Age Groups by Gender")
plt.show()
```

```python
#total amount as age group
sales_age = df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sales_age
```

```python
#top 10 states by orders
sales_st = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values('Orders' , ascending = False).head(10)
sales_st
```

```python
sns.set(rc = {'figure.figsize' : (15,8)})
sns.barplot(data = sales_st , x ='State', y='Orders')
```

```python
#total states by amount and sales
sales_st = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values('Amount' , ascending = False).head(10)
sales_st
sns.set(rc = {'figure.figsize' : (15,8)})
sns.barplot(data = sales_st , x ='State', y='Amount')
```

```python
ax = sns.countplot(x='Marital_Status', data =df)

sns.set(rc = {'figure.figsize': (8,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total states by amount and sales
marital = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values('Amount' , ascending = False).head(10)

ax = sns.barplot(x='Marital_Status', y ='Amount', hue ='Gender', data = marital)
sns.set(rc = {'figure.figsize': (8,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x='Occupation', data =df ,palette="Set2")

sns.set(rc = {'figure.figsize': (15,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total states by amount and sales
occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values('Amount' , ascending = False).head(10)

ax = sns.barplot(x='Occupation', y ='Amount', data = occupation , palette ='Set2')
sns.set(rc = {'figure.figsize': (15,8)})
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f')
    
```
